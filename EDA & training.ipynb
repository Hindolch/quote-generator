```python
data = open('data2.txt', 'r').read()
for i in data.split(', '):
    print(i)
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
tokenizer = Tokenizer()
```

```python
for i in data.split(', '):
    tokenizer.fit_on_texts([i])
```

```python
tokenizer.word_index
```

```python
VOCAB_SIZE = len(tokenizer.word_index)
```

```python
VOCAB_SIZE
```

```python
input_seq = []
for sent in data.split('\n'):
    tokenized_sent = tokenizer.texts_to_sequences([sent])[0]
    for i in range(1, len(tokenized_sent)):
        ngram = tokenized_sent[:i+1]
        input_seq.append(ngram)
input_seq
```

```python
MAX_INPUT_LEN = max(len(x) for x in input_seq)
MAX_INPUT_LEN
```

```python
from tensorflow.keras.utils import pad_sequences
padded_input = pad_sequences(input_seq, maxlen=MAX_INPUT_LEN, padding='pre')
padded_input
```

```python
X = padded_input[:,:-1]
X
```

```python
y = padded_input[:,-1]
y
```

```python
from tensorflow.keras.utils import to_categorical
y = to_categorical(y, num_classes=VOCAB_SIZE+1)
y
```

```python
print(X.shape)
print(y.shape)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dropout, Embedding, Dense
model = Sequential()
model.add(Embedding(713, 120, input_length=52))
model.add(LSTM(160))
model.add(Dropout(0.2))
model.add(Dense(713, activation='softmax'))
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(X,y, epochs=110)
```

```python
import numpy as np
text = 'she'
max_length = 100
for i in range(25):
    token = tokenizer.texts_to_sequences([text])[0]
    pad = pad_sequences([token], maxlen=53,padding='pre')
    pred = np.argmax(model.predict(pad))
    for word,index in tokenizer.word_index.items():
        if index == pred:
            text = text + ' '+ word
print(text)
```
